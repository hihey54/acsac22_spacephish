```python
import requests
import time
import datetime
import hashlib
from os import listdir
from os.path import isfile, join
```

```python
# request to API

token = "f06488eb3d0ff5bf0d6c2cb7fc68af33" # get yours from www.mlsec.io
url = 'https://api.mlsec.io/api/phish/submit_sample_all/?api_token='+token
headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
}
```

```python
def better_output(response, full=False, table=True):
    import re
    res = response.content.decode("utf-8")
    if full==True:
        print(res)
    res = res.split(",")
    d1 = re.findall(": (.*)", res[1])[0]
    d2 = re.findall(": (.*)",  res[2])[0]
    d3 = re.findall(": (.*)", res[3])[0]
    d4 = re.findall(": (.*)", res[4])[0]
    d5 = re.findall(": (.*)", res[5])[0]
    d6 = re.findall(": (.*)", res[6])[0]
    d7 = re.findall(": (.*)", res[7])[0]
    d8 = re.findall(": (.*)", res[8])[0]
    avg = (float(d1) + float(d2) + float(d3) + float(d4) + float(d5) + float(d6) + float(d7) + float(d8))/8
    if table==True:
        print(d1)
        print(d2)
        print(d3)
        print(d4)
        print(d5)
        print(d6)
        print(d7)
        print(d8)
        print("AVG: ", avg)
    
```

```python
# Check original:

now = datetime.datetime.now()
print ("Current date and time: ")
print (now.strftime("%Y-%m-%d %H:%M:%S"))


input_folder = "./data/"
onlyfiles = [f for f in listdir(input_folder) if isfile(join(input_folder, f))]

for page in onlyfiles:
    input_file = input_folder + page
    print(input_file)
    sha256_hash = hashlib.sha256()
    with open(input_file,"rb") as f:
        # Read and update hash string value in blocks of 4K
        for byte_block in iter(lambda: f.read(4096),b""):
            sha256_hash.update(byte_block)
        print(sha256_hash.hexdigest())
    with open(input_file, 'rb') as f:
        original_data = f.read()#.replace(b'\n', b'')
        original_response = requests.post(url, headers=headers, data=original_data)
        better_output(original_response)

        
now = datetime.datetime.now()
print ("Current date and time: ")
print (now.strftime("%Y-%m-%d %H:%M:%S"))
```
